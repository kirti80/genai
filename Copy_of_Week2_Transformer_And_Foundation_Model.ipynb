Transformer Architecture:
- Transformer models are neural networks
- It has high-level architecture which uses an encoder and decoder
- The encoder takes the input data and is able to extract meaning and context from the data and pass onto the decoder
- This is performed by converting the input sequence into numbers called embeddings
- Embedding is performred by mechanism called **Self Attention**,sent through  feed forward network,through normalization layer.
- Self-Attention mechanism captures importance of each word and the relationships between the words.
- Example,each person pays attention to everyone else's ideas. They consider the relevance, importance of each party idea and how the ideas all relate to each other. This happens as part of a brainstorming and collaboration session.

Training transformers involves two steps:
- pretraining
   - Trained on large amounts of raw (text) data, with the internet as the primary data source.
   - The training is done using self-supervised learning, an innovative type of training that doesn‚Äôt require human action to label data.
   - The goal of **pre-training is to learn the statistical patterns of the language.**
- fine-tuning
  - **Trained on a narrower, domain-specific dataset **generated with the help of human reviewers following certain guidelines.

Foundation Models:
- Trained on massive datasets
- Large deep learning neural networks that have changed the way data scientists approach machine learning (ML)
- Rather than develop artificial intelligence (AI) from scratch, data scientists use a foundation model as a starting point to develop ML models that power new applications more quickly and cost-effectively.
- Trained on a broad spectrum of generalized and unlabeled data and capable of performing a wide variety of general tasks such as understanding language, generating text and images, and conversing in natural language.
- Use as base models for developing more specialized downstream applications.

#Tuning and deploy a foundation model

Tune foundation models:
- Model tuning works by providing a model with a training dataset that contains many examples of a unique task
- For unique, niche tasks, you can get significant improvements in model performance by tuning the model on a modest number of examples.
- After you tune a model, fewer examples are required in its prompts.
- Model tunning Methods supported by VertexAI:
  - Supervised tuning
  - Reinforcement learning from human feedback (RLHF) tuning
  - Tuning and distillation

Supervised Tuning:
- Improves the performance of a model by teaching it a new skill
- Data that contains hundreds of labeled examples is used to teach the model to mimic a desired behavior or task
- The output of the tuning job is a new model that combines the newly learned parameters with the original model

Models that support supervised tuning:
- Text generation - text-bison@002 and text-bison-32k
- Text chat - chat-bison@002 and chat-bison-32k
- Code generation - code-bison@002 and code-bison-32k
- Code chat - codechat-bison@002 and codechat-bison-32k
- Text embeddings - textembedding-gecko@001


Workflow for supervised model tuning:
- Prepare your model tuning dataset.
- Upload the model tuning dataset to a Cloud Storage bucket.
- Create a supervised model tuning job.

```python
vaccine_url="https://www.kaggle.com/datasets/gauravduttakiit/vaccines-sentiment-analysis"
od.download(vaccine_url,force=True)
```

```python
from pathlib import Path
import pandas as pd
import torch
pd.set_option('display.max_colwidth', 1)
from torch.utils.data import Dataset, DataLoader
def read_vaccine_split(split_dir):
    split_dir = Path(split_dir)
    texts = []
    labels = []
    for label_dir in ['-1.0','0.0','1.0']:
        for text_file in (split_dir/label_dir).iterdir():
            texts.append(text_file.read_text())
            if  label_dir == "0.0":
                labels.append(0.0 )
            elif  label_dir == "-1.0":
                labels.append(-1.0)
            else :
                labels.append(1.0)

    return texts, labels

train_texts, train_labels = read_vaccine_split('/content/vaccines-sentiment-analysis/train')
test_texts, test_labels = read_vaccine_split('/content/vaccines-sentiment-analysis/test')
```

```python
train=pd.DataFrame(list(zip(train_texts,train_labels)),columns=["text",'label'])
```

```python
def to_sentiment(label):
    label = int(label)
    if label == -1:
      return "Neutral"
    elif label == 0:
      return "Positive"
    else:
      return "Negative"
```

```python
train['label']=train['label'].astype(int)
train['sentiment'] = train.label.apply(to_sentiment)
train=train.drop(columns='label')
# train['label']=train['sentiment']
# train=train.drop(columns='sentiment')
```

Tune the model:

- Next, you tune the model using the `tune_model()` method, with the following parameters:

 `training_data`: A pandas Dataframe or Cloud Storage location of the training data for tuning the model.<br>
`learning_rate_multiplier`: A multiplier to apply to the recommended learning rate. To use the recommended learning rate, use 1.0. <br>
`train_steps`: The number of steps to run for model tuning. The batch size varies by tuning location:<br>
- us-central1 has a batch size of 8.
- europe-west4 has a batch size of 24.<br>

If there are 240 examples in a training dataset, in europe-west4, it takes 240 / 24 = 10 steps to process the entire dataset once. In us-central1, it takes 240 / 8 = 30 steps to process the entire dataset once. The default value is 300.<br>

`tuning_job_location`: The region where the tuning job should be run. Supported regions are: `us-central1` and `europe-west4`.<br>
`tuned_model_location`: The region where the tuned model should be deployed.
