```python
from pathlib import Path
import pandas as pd
def read_imdb_split(split_dir):
    split_dir = Path(split_dir)
    texts = []
    labels = []
    for label_dir in ["pos", "neg"]:
        for text_file in (split_dir/label_dir).iterdir():
            texts.append(text_file.read_text())
            labels.append(0 if label_dir is "neg" else 1)

    return texts, labels

train_texts, train_labels = read_imdb_split('/Users/kirtikamerkar/Downloads/aclImdb/train')
test_texts, test_labels = read_imdb_split('/Users/kirtikamerkar/Downloads/aclImdb/test')
```
